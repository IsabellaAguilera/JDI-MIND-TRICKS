```python
import pandas as pd
import numpy as np
```

```python
#Cleaning the zillow data using 11 datasets, with each dataset representing a different rental type (i.e. Studio, 2-bedroom)
one_bed_url='1_bed.csv'
two_bed_url='2_bed.csv'

#Read csv file
one_bed_df=pd.read_csv(one_bed_url)
one_bed_df.head()
```

```python
#Filter out all states except for NJ
nj_one_bed_df=one_bed_df[one_bed_df["State"] == "NJ"]
nj_one_bed_df.head()
```

```python
#Remove unnecessary columns
nj_one_bed_df=nj_one_bed_df.drop(nj_one_bed_df.loc[:,nj_one_bed_df.columns.str.startswith('StateCodeFIPS')],axis=1)
nj_one_bed_df=nj_one_bed_df.drop(nj_one_bed_df.loc[:,nj_one_bed_df.columns.str.startswith('MunicipalCodeFIPS')],axis=1)
nj_one_bed_df=nj_one_bed_df.drop(nj_one_bed_df.loc[:,nj_one_bed_df.columns.str.startswith('SizeRank')],axis=1)
nj_one_bed_df=nj_one_bed_df.drop(nj_one_bed_df.loc[:,nj_one_bed_df.columns.str.startswith('2010')],axis=1)
nj_one_bed_df=nj_one_bed_df.drop(nj_one_bed_df.loc[:,nj_one_bed_df.columns.str.startswith('2011')],axis=1)
nj_one_bed_df=nj_one_bed_df.drop(nj_one_bed_df.loc[:,nj_one_bed_df.columns.str.startswith('2012')],axis=1)
nj_one_bed_df=nj_one_bed_df.drop(nj_one_bed_df.loc[:,nj_one_bed_df.columns.str.startswith('2019')],axis=1)

nj_one_bed=nj_one_bed_df.reset_index()

del nj_one_bed['index']

#Fill NaN values with 0, which will later be replaced with NaN values once again to find mean of all the months
#of each year
nj_one_bed=nj_one_bed.fillna(0)

nj_one_bed
```

```python
#Create a new dataframe with all the months of one year to calculate the average of each row
one_bed_2013=pd.DataFrame(nj_one_bed[["2013-01","2013-02","2013-03","2013-04","2013-05","2013-06","2013-07","2013-08",
                                      "2013-09","2013-10","2013-11","2013-12"]])
                                      
                                      
                            
#Calculate the mean of each row
one_bed_2013['2013 Mean']=one_bed_2013.mean(axis=1)
one_bed_2013 = one_bed_2013.replace(0, np.NaN)

one_bed_2013
```

```python
one_bed_2014=pd.DataFrame(nj_one_bed[["2014-01","2014-02","2014-03","2014-04","2014-05","2014-06","2014-07","2014-08",
                                      "2014-09","2014-10","2014-11","2014-12"]])
                                      
                                      
                            

one_bed_2014['2014 Mean']=one_bed_2014.mean(axis=1)
one_bed_2014 = one_bed_2014.replace(0, np.NaN)

one_bed_2014
```

```python
one_bed_2015=pd.DataFrame(nj_one_bed[["2015-01","2015-02","2015-03","2015-04","2015-05","2015-06","2015-07","2015-08",
                                      "2015-09","2015-10","2015-11","2015-12"]])
                                      
                                      
                            

one_bed_2015['2015 Mean']=one_bed_2015.mean(axis=1)
one_bed_2015 = one_bed_2015.replace(0, np.NaN)

one_bed_2015
```

```python
one_bed_2016=pd.DataFrame(nj_one_bed[["2016-01","2016-02","2016-03","2016-04","2016-05","2016-06","2016-07","2016-08",
                                      "2016-09","2016-10","2016-11","2016-12"]])
                                      
                                      
                            

one_bed_2016['2016 Mean']=one_bed_2016.mean(axis=1)
one_bed_2016 = one_bed_2016.replace(0, np.NaN)

one_bed_2016
```

```python
one_bed_2017=pd.DataFrame(nj_one_bed[["2017-01","2017-02","2017-03","2017-04","2017-05","2017-06","2017-07","2017-08",
                                      "2017-09","2017-10","2017-11","2017-12"]])
                                      
                                      
                            

one_bed_2017['2017 Mean']=one_bed_2017.mean(axis=1)
one_bed_2017 = one_bed_2017.replace(0, np.NaN)

one_bed_2017
```

```python
one_bed_2018=pd.DataFrame(nj_one_bed[["2018-01","2018-02","2018-03","2018-04","2018-05","2018-06","2018-07","2018-08",
                                      "2018-09","2018-10","2018-11","2018-12"]])
                                      
                                      
                            

one_bed_2018['2018 Mean']=one_bed_2018.mean(axis=1)
one_bed_2018 = one_bed_2018.replace(0, np.NaN)

one_bed_2018
```

```python
#Create a master dataframe with the averages of each year
one_bed_master=pd.DataFrame(nj_one_bed[["RegionName","State","Metro"]])
one_bed_master["2013 Average"]=one_bed_2013["2013 Mean"]
one_bed_master["2014 Average"]=one_bed_2014["2014 Mean"]
one_bed_master["2015 Average"]=one_bed_2015["2015 Mean"]
one_bed_master["2016 Average"]=one_bed_2016["2016 Mean"]
one_bed_master["2017 Average"]=one_bed_2017["2017 Mean"]
one_bed_master["2018 Average"]=one_bed_2018["2018 Mean"]


one_bed_master
```

```python
#Create csv file for master dataframe
one_bed_master.to_csv('1_bed_master.csv')
```

```python
nj_all_homes_df=nj_all_homes_df.drop(nj_all_homes_df.loc[:,nj_all_homes_df.columns.str.startswith('StateCodeFIPS')],axis=1)
nj_all_homes_df=nj_all_homes_df.drop(nj_all_homes_df.loc[:,nj_all_homes_df.columns.str.startswith('MunicipalCodeFIPS')],axis=1)
nj_all_homes_df=nj_all_homes_df.drop(nj_all_homes_df.loc[:,nj_all_homes_df.columns.str.startswith('SizeRank')],axis=1)
nj_all_homes_df=nj_all_homes_df.drop(nj_all_homes_df.loc[:,nj_all_homes_df.columns.str.startswith('2010')],axis=1)
nj_all_homes_df=nj_all_homes_df.drop(nj_all_homes_df.loc[:,nj_all_homes_df.columns.str.startswith('2011')],axis=1)
nj_all_homes_df=nj_all_homes_df.drop(nj_all_homes_df.loc[:,nj_all_homes_df.columns.str.startswith('2012')],axis=1)
nj_all_homes_df=nj_all_homes_df.drop(nj_all_homes_df.loc[:,nj_all_homes_df.columns.str.startswith('2019')],axis=1)


nj_all_homes=nj_all_homes_df.reset_index()

del nj_all_homes['index']
nj_all_homes=nj_all_homes.fillna(0)

nj_all_homes
```

```python
all_homes_2013=pd.DataFrame(nj_all_homes[["2013-01","2013-02","2013-03","2013-04","2013-05","2013-06","2013-07","2013-08",
                                      "2013-09","2013-10","2013-11","2013-12"]])
                                      
                                      
                            

all_homes_2013['2013 Mean']=all_homes_2013.mean(axis=1)
all_homes_2013 = all_homes_2013.replace(0, np.NaN)

all_homes_2013
```

```python
all_homes_2014=pd.DataFrame(nj_all_homes[["2014-01","2014-02","2014-03","2014-04","2014-05","2014-06","2014-07","2014-08",
                                      "2014-09","2014-10","2014-11","2014-12"]])
                                      
                                      
                            

all_homes_2014['2014 Mean']=all_homes_2014.mean(axis=1)
all_homes_2014 = all_homes_2014.replace(0, np.NaN)

all_homes_2014
```

```python
all_homes_master=pd.DataFrame(nj_all_homes[["RegionName","State","Metro"]])
all_homes_master["2013 Average"]=all_homes_2013["2013 Mean"]
all_homes_master["2014 Average"]=all_homes_2014["2014 Mean"]
all_homes_master["2015 Average"]=all_homes_2015["2015 Mean"]
all_homes_master["2016 Average"]=all_homes_2016["2016 Mean"]
all_homes_master["2017 Average"]=all_homes_2017["2017 Mean"]
all_homes_master["2018 Average"]=all_homes_2018["2018 Mean"]


all_homes_master
```
