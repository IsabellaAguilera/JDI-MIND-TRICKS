```python
import pandas as pd
```

```python
one_bed_url='1_bed.csv'
two_bed_url='2_bed.csv'

one_bed_df=pd.read_csv(one_bed_url)
one_bed_df.head()
```

```python
two_bed_df=pd.read_csv(two_bed_url)
two_bed_df.head()
```

```python
#1-bed and 2-bed merge
one_two_bed_merge = pd.merge(one_bed_df, two_bed_df, on="RegionName", how="outer")
one_two_bed_merge
```
