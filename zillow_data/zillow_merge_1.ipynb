```python
import pandas as pd
import numpy as np
```

```python
#Merge the master dataframe of each rental type to create a dataframe with all the rental types
one_bed_url='1_bed_master.csv'
two_bed_url='2_bed_master.csv'

#Read csv 1
one_bed_df=pd.read_csv(one_bed_url)
one_bed_df.head()
```

```python
#Read csv 2
two_bed_df=pd.read_csv(two_bed_url)
two_bed_df.head()
```

```python
#Outer merge based on county ("RegionName")
one_two_bed_merge = pd.merge(one_bed_df, two_bed_df, on="RegionName", how="outer")
one_two_bed_merge.head()
```

```python
#Clean up the columns by renaming the x and y columns
one_two_bed_merge=one_two_bed_merge.rename(columns={"State_x":"State","Metro_x":"Metro","2013 Average (One Bedroom)":
                                                    "2013 Average (One Bed)","2014 Average_x":"2014 Average (One Bed)",
                                                   "2015 Average_x":"2015 Average (One Bed)","2016 Average_x":
                                                   "2016 Average (One Bed)", "2017 Average_x":"2017 Average (One Bed)",
                                                   "2018 Average_x":"2018 Average (One Bed)","2013 Average_y":
                                                   "2013 Average (Two Bed)","2014 Average_y":"2014 Average (Two Bed)",
                                                   "2015 Average_y":"2015 Average (Two Bed)","2016 Average_y":
                                                   "2016 Average (Two Bed)","2017 Average_y":"2017 Average (Two Bed)",
                                                   "2018 Average_y":"2018 Average (Two Bed)"})




one_two_bed_merge.head()
```

```python
#3rd merge
three_bed_url='3_bed_master.csv'

three_bed_df=pd.read_csv(three_bed_url)
three_bed_df.head()
```

```python
three_bed_merge = pd.merge(one_two_bed_merge, three_bed_df, on="RegionName", how="outer")


del (three_bed_merge["Unnamed: 0"],three_bed_merge["State_y"],three_bed_merge["Metro_y"])
three_bed_merge.head()
```

```python
three_bed_merge=three_bed_merge.rename(columns={"State_x":"State","Metro_x":"Metro","2013 Average":"2013 Average (Three Bed)",
                                                "2014 Average":"2014 Average (Three Bed)","2015 Average":"2015 Average (Three Bed)",
                                                "2016 Average":"2016 Average (Three Bed)","2017 Average":"2017 Average (Three Bed)",
                                                "2018 Average":"2018 Average (Three Bed)"
                                                })
three_bed_merge
```

```python
four_bed_url='4_bed_master.csv'

four_bed_df=pd.read_csv(four_bed_url)

del four_bed_df["Unnamed: 0"]
four_bed_df.head()
```

```python
four_bed_merge = pd.merge(three_bed_merge, four_bed_df, on="RegionName", how="outer")

del (four_bed_merge["State_y"],four_bed_merge["Metro_y"])
four_bed_merge.head()
```

```python
four_bed_merge=four_bed_merge.rename(columns={"State_x":"State","Metro_x":"Metro","2013 Average":"2013 Average (Four Bed)",
                                              "2014 Average":"2014 Average (Four Bed)","2015 Average":"2015 Average (Four Bed)",
                                              "2016 Average":"2016 Average (Four Bed)","2017 Average":"2017 Average (Four Bed)",
                                              "2018 Average":"2018 Average (Four Bed)"
                                                })
four_bed_merge
```

```python
five_bed_url='5_bed_master.csv'

five_bed_df=pd.read_csv(five_bed_url)

del five_bed_df["Unnamed: 0"]
five_bed_df
```

```python
five_bed_merge = pd.merge(four_bed_merge, five_bed_df, on="RegionName", how="outer")

del (five_bed_merge["State_y"],five_bed_merge["Metro_y"])

five_bed_merge.head()
```

```python
five_bed_merge=five_bed_merge.rename(columns={"State_x":"State","Metro_x":"Metro","2013 Average":"2013 Average (Five Bed)",
                                              "2014 Average":"2014 Average (Five Bed)","2015 Average":"2015 Average (Five Bed)",
                                              "2016 Average":"2016 Average (Five Bed)","2017 Average":"2017 Average (Five Bed)",
                                              "2018 Average":"2018 Average (FIve Bed)"
                                            })

five_bed_merge
```

```python
#Create csv file for the 1 to 5 bedroom merged dataframe
five_bed_merge.to_csv('1_to_5_bed_merged.csv')
```
