```python
max_rent = {}
for x in range(3,len(master_rentals_df.columns)):
    for x in master_rentals_df.iloc[:, [x]]:
        i = master_rentals_df['County'].iloc[master_rentals_df[x].idxmax()]
        y = master_rentals_df[x].iloc[master_rentals_df[x].idxmax()]
        max_rent[x] = i, y
max_rent
```

```python
min_rent = {}
for x in range(3,len(master_rentals_df.columns)):
    for x in master_rentals_df.iloc[:, [x]]:
        i = master_rentals_df['County'].iloc[master_rentals_df[x].idxmin()]
        y = master_rentals_df[x].iloc[master_rentals_df[x].idxmin()]
        min_rent[x] = i, y
min_rent
```

```python
answer = 'yes'
while answer == 'yes':
    question = input('Want to know which counties had the highest or lowest rental prices in New Jersey between 2013 - 2018?' )
    min_max = input('Highest or lowest rent prices? ')
    rental = input('You can choose from the following: One Bed, Two Bed, Studio, and All Homes. Which type of rental are you looking for? ')
    year = str(input('In what year? '))

    try:
        countymax = max_rent[f'{year} Average ({rental})'][0]
        countymin = min_rent[f'{year} Average ({rental})'][0]
        pricemax = str(round(max_rent[f'{year} Average ({rental})'][1],0))
        pricemin = str(round(min_rent[f'{year} Average ({rental})'][1],0))

        if min_max  == 'highest':
             print("The highest rental average for a " + rental + " in " + year + " is $" + pricemax + " in " + countymax + ", New Jersey." )
        else:
             print("The lowest rental average for a " + rental + " in " + year + " is $" + pricemin + " in " + countymin + ", New Jersey.")
    except KeyError:
        print('Oops, need to capitalize bed')
    
    answer = input("Would you like to make try again? ('yes' or 'no') ")
```
