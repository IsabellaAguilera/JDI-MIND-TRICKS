```python
url_1='Average_income_file_nj.csv'
income=pd.read_csv(url_1)
income

url_2='zillow_data/1_bed_master.csv'
one_bed=pd.read_csv(url_2)
one_bed

```

```python
income=income.rename(columns={"Area_name":"RegionName"})

income.columns=income.columns.str.replace('County, NJ','County')
income['RegionName'] = income['RegionName'].str.split(",", n=1,expand=True)
income
```

```python
one_bed_income = pd.merge(income, one_bed, on="RegionName",how="inner")

pd.options.display.float_format = '{:,.2f}'.format
pd.set_option('display.max_columns', 500)
pd.set_option('display.width', 1000)

one_bed_income
```

```python
income_clean=pd.DataFrame(one_bed_income[["Median_Household_Income_2013","Median_Household_Income_2014",
                                          "Median_Household_Income_2015","Median_Household_Income_2016",
                                      "Median_Household_Income_2017","Median_Household_Income_2018"]])

pd.options.display.float_format = '{:,.2f}'.format

income_clean
```

```python
one_bed_clean=pd.DataFrame(one_bed_income[["2013 Average","2014 Average","2015 Average","2016 Average","2017 Average",
                                          "2018 Average"]])
one_bed_clean
```

```python
years=[2013,2014,2015,2016,2017,2018]

plt.plot(years, one_bed_hudson, marker ='o', color='blue', label="Hudson County")
plt.plot(years, one_bed_essex, marker ='o', color='red', label="Essex County")
plt.plot(years, one_bed_passaic, marker ='o', color='green', label="Passaic County")
plt.plot(years, one_bed_monmouth, marker ='o', color='purple', label="Monmouth County")




plt.legend(loc="upper left",prop={'size':8})
plt.ylim(0, 2500)

plt.xlim(2012,2018)
plt.title("One Bed Rental Averages Over 6 Years")
plt.xlabel("Years")
plt.ylabel("Average Rental Values in Dollars")
```
