```python
import pandas as pd
import numpy as np
import random
import matplotlib.pyplot as plt
```

```python
#Read csv file for average income with pandas
url_1='Average_income_nj.csv'
income=pd.read_csv(url_1)
income
```

```python
#Read csv file for one bed rentals with pandas
url_2='../zillow_data/1_bed_master.csv'
one_bed=pd.read_csv(url_2)
one_bed
```

```python
#Change "Area_name" column to "RegionName" so that it matches the header for counties in the one_bed dataset
#to prepare for merge
income=income.rename(columns={"Area_name":"RegionName"})

#Drop ",NJ" by splitting so that only the county is named, not the state
#Again, this is to prepare for the merge so that the county values match
income.columns=income.columns.str.replace('County, NJ','County')
income['RegionName'] = income['RegionName'].str.split(",", n=1,expand=True)
income
```

```python
#Inner merge one_bed dataset with income dataset on county("RegionName") so that we only keep rows for counties
#with values in both datasets
one_bed_income = pd.merge(income, one_bed, on="RegionName",how="inner")

pd.options.display.float_format = '{:,.2f}'.format
pd.set_option('display.max_columns', 500)
pd.set_option('display.width', 1000)

one_bed_income
```

```python
#Create new dataframe with just the annual median household income as columns to later create lists from each row
#to plot
income_clean=pd.DataFrame(one_bed_income[["Median_Household_Income_2013","Median_Household_Income_2014",
                                          "Median_Household_Income_2015","Median_Household_Income_2016",
                                      "Median_Household_Income_2017","Median_Household_Income_2018"]])

pd.options.display.float_format = '{:,.2f}'.format

income_clean
```
